#### Imports

```python
import time

from stable_baselines3 import PPO

from sacleric01 import SA_CLERIC

from ma_v03 import MA_PARTY
```

#### Testing SA_CLERIC

```python
env=SA_CLERIC(render_mode="human") #, debug_mode=True)
reward_total = 0
all_heroes_alive = 0

runs = 1000

for run in range(1, runs):
    print(f"Run {run}")

    observations, infos = env.reset()
    termination = False

    while not termination:    
        action = env.action_space.sample(infos["action_mask"])
        
        observations, reward, termination, dummy, infos = env.step(action)

        reward_total += reward

        if reward == 100:
            all_heroes_alive += 1
    env.close()

reward_per_round = reward_total / runs
all_heroes_alive_average = all_heroes_alive / runs * 100
print(f"Reward per round: {reward_per_round}") # 73.27
print(f"Success rate: {all_heroes_alive_average} %") # 61.7 %
```

#### Train a SA_Fighter Model

```python
env=SA_CLERIC()
model = PPO("MlpPolicy", env, verbose=1, tensorboard_log="./ppo_sa_cleric_tensorboard/")
model.learn(total_timesteps=100_000) # 400000

model_name = f"./sa_cleric_models/{env.unwrapped.metadata.get('name')}_{time.strftime('%Y%m%d-%H%M%S')}"
model.save(model_name)

print("Model has been saved.")
print(f"Finished training on {str(env.unwrapped.metadata['name'])}.")
print(model_name)
```
